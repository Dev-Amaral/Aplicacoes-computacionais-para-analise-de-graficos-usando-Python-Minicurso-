# 1. matplotlib.pyplot.plot()

Esse método é o mais comum para criar gráficos de funções contínuas, como lineares, quadráticas, exponenciais, etc.

Ele é usado para plotar gráficos 2D simples de x versus y.

Exemplo: plotando um função quadrática f(x)=x^2



```python
import matplotlib.pyplot as plt
import numpy as np

# Dados para o gráfico
x = np.linspace(-6, 6, 40)
y = x**3 + 4**x
z = np.exp(x)
w = 2*x + x**2

plt.plot(x, y, label='$ f(x) = x^3 + 4^x $')
plt.plot(x, z, label='$ g(x) = e^x $')
plt.plot(x, w, label='$ h(x) = 2*x + x^2$')
plt.legend()
plt.show() # no colab não há a necessidade
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados para o gráfico
x = np.linspace(0, 12, 100)  # cria 100 pontos entre -10 e 10
y = x**2

plt.plot(x, y, label="$f(x) = x^2$")
plt.grid()
plt.title('faturamento versus mês')
plt.ylabel('faturamento (R$)')
plt.xlabel('meses do ano')
plt.legend()
plt.show()
```

# 3. matplotlib.pyplot.bar()

Este método é utilizado para criar gráficos de barras, mas pode ser útil se você estiver comparando valores discretos em diferentes categorias.

Exemplo: Criando um gráfico de barras para valores de uma função modular:

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados para o gráfico
x = np.array([1, 2, 3, 4, 5])
y = np.abs(x-5)

plt.bar(x, y, label='salarios por ranking da empresa', width=0.8, edgecolor='red')
plt.xlabel('Funcionário rankeado')
plt.ylabel('Salário minimo x y')
plt.legend()
plt.show()
```

# 4. matplotlib.pyplot.loglog()

Usado para gráficos onde ambos os eixos estão em escala logarítmica. Ideal para plotar funções exponenciais ou logarítmicas.

Exemplo: Plotando uma função logarítmica
𝑓(𝑥) = log(𝑥):

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados para o gráfico
x = np.linspace(1, 100, 100)
y = np.exp(x)
w = x+2 #podemos plotar esta função no mesmo gráfico

# Criando o nosso gráfico
plt.plot(x, y, label="y = log(x)")
#plt.plot(x,w, label="w=x+2")
plt.show()
```

# 5. Adicionando legendas aos nossos gráficos

Percebam que todos os nossos gráficos até agora eram carentes de informações para aqueles que buscam uma análise.

Para resolver este impasse vamos aprender como adicionar informações às nossas imagens.


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = 2**x
z = x ** 2

'''
Então como poderiamos por exemplo criar o gráfico utilizando o conjunto de dados acima, e colocarmos informações de:
1. quais grandezas relacionadas ao eixo x e y
2. colocar uma legenda indicando qual função gerou esta curva
3. Definir um título para nosso gráfico
4. E para melhorar nossa visualização vamos colocar um grid ?
'''

plt.plot(x, y, label='$ P(t) = 2^t $')
plt.plot(x, z, label='$ P(t) = t^2 $', color="red")
plt.xlabel('Tempo (min)') # legenda do eixo x
plt.ylabel('Número de indivíduos') # legenda do eixo y
#plt.title('Evolução da População de bactérias') # titulo do gráfico
plt.grid() # determina uma malha quadriculada
plt.legend() # mostra as legendas
plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = 2**x
z = 10*x
plt.plot(x, y, label='desgaste do motor modelo 1 em 4 anos')
plt.plot(x, z, label='desgaste do motor modelo 2 em 4 anos')
plt.legend()
plt.show()
```

# 6. Gráficos interativos em 3D

```python
!pip install plotly==5.15.0
import plotly.graph_objects as go
import numpy as np

# Função a ser plotada
def f(x, y):
  return np.sin(np.sqrt(x**2 + y**2))

# Domínio da função
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Cria o gráfico de superfície
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])

# Configura o layout do gráfico
fig.update_layout(title='Gráfico 3D Interativo', autosize=False,
                  width=800, height=800,
                  margin=dict(l=65, r=50, b=65, t=90))

# Exibe o gráfico
fig.show()
```
